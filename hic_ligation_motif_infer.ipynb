# Hi-C ligation motif infer
## What this nootbook for?
> This notbook is for [ALLHiC](https://github.com/tangerzhang/ALLHiC) to interfer what kind of HiC enzymes ligation motif you should use when your data generated by arima four enzyme cocktail toolkit in the step of ALLHiC_partition

```python
#!/usr/bin/env python

enzymes = []
number = 0
times = 0
vaild = ['A','T','C','G','N','^']
five_prime = []
three_prime = []
final_result = []
while True:
    number = input("How many enzymes used in your Hi-C library constraction?: ")
    if number.isdigit():
        number = int(number)
    else:
        print("Please only input a  non-negative real numbers here.")
        break
    if number >= 0:
        while times < number:
            input_enzyme = input("Please enter the first enzyme, you can only use A T C G N and ^ to represent the cut site: ")
            for nucl in input_enzyme:
                if nucl in vaild:
                    if "^" not in input_enzyme:
                        print("You should use '^' to define the cut site.")
                        break
                    else:
                        pass
                else:
                    print("There are looks like you inputed something not only A T C G N ^, please check it!")
                    break
            enzymes.append(input_enzyme)
            times += 1
        print("The enzymes you inputed are {}".format(enzymes))
        break
    elif number <= 0:
        print("You can only input non-negative real numbers here. Please check it again.")
        break
    else:
        print("You should enter a number here, please check it again.")
        break


#enzymes = ['T^TAA', 'C^TNAG', 'G^ANTC', '^GATC']


for enzyme in enzymes:
    if enzyme.startswith("^",0,1):
        site = enzyme.replace('^','')
        five_prime.append(site)
        three_prime.append(site)
        print("Due to this enzyme {} split at first site, the 5' of it is: {} and 3' of it is: {}".format(enzyme,site, site))
    elif enzyme.startswith("^",1,2):
        site = enzyme.replace('^','')
        five = site[0:-1]
        three = site[1:]
        five_prime.append(five)
        three_prime.append(three)
        print("Due to this enzyme {} split at second site, the 5' of it is: {} and 3' of it is: {}".format(enzyme,five, three))

print("The final result:")
for i in five_prime:
    for j in three_prime:
        print(i+j,end=",")
```
